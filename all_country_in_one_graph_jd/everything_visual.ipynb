```python
# Dependencies
from matplotlib import pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
import json
```

```python
# reading the csv
ca = pd.read_csv("CAvideos.csv", encoding="utf-8")
de = pd.read_csv("DEvideos.csv", encoding="utf-8")
fr = pd.read_csv("FRvideos.csv", encoding="utf-8")
gb = pd.read_csv("GBvideos.csv", encoding="utf-8")
ind = pd.read_csv("INvideos.csv", encoding="utf-8")
jp = pd.read_csv("JPvideos.csv", encoding="utf-8")
kr = pd.read_csv("KRvideos.csv", encoding="utf-8")
mx = pd.read_csv("MXvideos.csv", encoding="utf-8")
ru = pd.read_csv("RUvideos.csv", encoding="utf-8")
us = pd.read_csv("USvideos.csv", encoding="utf-8")

ca = ca.dropna(how="any")
de = de.dropna(how="any")
fr = fr.dropna(how="any")
gb = gb.dropna(how="any")
ind = ind.dropna(how="any")
jp = jp.dropna(how="any")
kr = kr.dropna(how="any")
mx = mx.dropna(how="any")
ru = ru.dropna(how="any")
us = us.dropna(how="any")

ca[["category_id", "views", "likes", "dislikes", "comment_count"]] = ca[["category_id", "views", "likes", "dislikes", "comment_count"]].astype(np.int64)
de[["category_id", "views", "likes", "dislikes", "comment_count"]] = de[["category_id", "views", "likes", "dislikes", "comment_count"]].astype(np.int64)
fr[["category_id", "views", "likes", "dislikes", "comment_count"]] = fr[["category_id", "views", "likes", "dislikes", "comment_count"]].astype(np.int64)
gb[["category_id", "views", "likes", "dislikes", "comment_count"]] = gb[["category_id", "views", "likes", "dislikes", "comment_count"]].astype(np.int64)
ind[["category_id", "views", "likes", "dislikes", "comment_count"]] = ind[["category_id", "views", "likes", "dislikes", "comment_count"]].astype(np.int64)
jp[["category_id", "views", "likes", "dislikes", "comment_count"]] = jp[["category_id", "views", "likes", "dislikes", "comment_count"]].astype(np.int64)
kr[["category_id", "views", "likes", "dislikes", "comment_count"]] = kr[["category_id", "views", "likes", "dislikes", "comment_count"]].astype(np.int64)
mx[["category_id", "views", "likes", "dislikes", "comment_count"]] = mx[["category_id", "views", "likes", "dislikes", "comment_count"]].astype(np.int64)
ru[["category_id", "views", "likes", "dislikes", "comment_count"]] = ru[["category_id", "views", "likes", "dislikes", "comment_count"]].astype(np.int64)
us[["category_id", "views", "likes", "dislikes", "comment_count"]] = us[["category_id", "views", "likes", "dislikes", "comment_count"]].astype(np.int64)
```
